```python
txt_path = './data/knowledgeTag_skillID.txt'

skill_dic = {}

with open(txt_path, "r") as file:
    lines = file.readlines()
    for line in lines:
        data = line.strip().split('\t')
        skill_dic[int(data[1])] = int(data[0])
```

```python
skill_dic
```

```python
import json

with open('./data/knowledgeTag_skillID.json', 'w', encoding='utf-8') as f:
    json.dump(skill_dic, f, ensure_ascii=False)
```
