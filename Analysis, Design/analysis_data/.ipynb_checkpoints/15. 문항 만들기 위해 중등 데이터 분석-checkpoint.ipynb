```python
# 그냥 리스트로 만들어서 이 파일 안에서 사용하기
import json

json_path = './data/knowledge_tags.json'
with open(json_path, 'r', encoding='utf-8') as j :
    knowledge_tags = json.load(j)
    
knowledge_space = []
for knowledge_tag_id in knowledge_tags :
    from_concept_id = knowledge_tags[knowledge_tag_id]['fromConcept']['id']
    to_concept_id = knowledge_tags[knowledge_tag_id]['toConcept']['id']
    knowledge_space.append([from_concept_id, to_concept_id])
```

```python
# 우선 mid로 한 번 시도
result = {}
for concept_id in mid :
    to_list = []
    for from_to in knowledge_space :
        if (from_to[0]==concept_id) :
            to_concept_id = from_to[1]
            to_to_dic = {}
            to_to_list = []
            for from_to_to in knowledge_space:
                if (from_to_to[0]==to_concept_id):
                    to_to_list.append(from_to_to[1])
            to_to_dic[to_concept_id]=to_to_list
            to_list.append(to_to_dic)
    result[concept_id]= to_list
```

```python
# 함수로 만들면
def find_to_to(arr, space):
    result = {}
    for concept_id in arr :
        to_list = []
        for from_to in space :
            if (from_to[0]==concept_id) :
                to_concept_id = from_to[1]
                to_to_dic = {}
                to_to_list = []
                for from_to_to in space:
                    if (from_to_to[0]==to_concept_id):
                        to_to_list.append(from_to_to[1])
                to_to_dic[to_concept_id]=to_to_list
                to_list.append(to_to_dic)
        result[concept_id]= to_list
    return result

def find_to_list(dic):
    depth0 = []
    depth1 = []
    depth2 = []
    for key, value in dic.items():
        depth0.append(key)
        for to in value:
            for k, v in to.items():
                depth1.append(k)
                for i in v:
                    depth2.append(i)
    return [depth0, depth1, depth2]
```
