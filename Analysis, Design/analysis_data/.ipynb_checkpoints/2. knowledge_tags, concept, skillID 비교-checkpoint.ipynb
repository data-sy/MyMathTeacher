```python
# set을 사용하려 했는데
# 파이썬에서는 딕셔너리를 집합의 원소로 직접 넣을 수 없습니다.
# 이것은 딕셔너리가 변경 가능한(mutable) 객체이기 때문입니다. 집합은 변경 불가능한(immutable) 객체만을 원소로 가질 수 있습니다.
```

```python
# 우선은 컨셉 갯수가 1865개가 맞는지 확인하려는 거라서
# 컨셉네임을 셋으로 추출해보자
```

```python
# concept name 추출
# 중복제거를 위해 set으로 만듦
concept_name_set = set()
for knowledge_tag in knowledge_tags.values() :
    concept_name_set.add(knowledge_tag['fromConcept']['name'])
    concept_name_set.add(knowledge_tag['toConcept']['name'])
print(len(concept_name_set))
```

```python
# concept id로 개수 세기
# concept name 추출
# 중복제거를 위해 set으로 만듦
concept_id_set = set()
for knowledge_tag in knowledge_tags.values() :
    concept_id_set.add(knowledge_tag['fromConcept']['id'])
    concept_id_set.add(knowledge_tag['toConcept']['id'])
print(len(concept_id_set))
```

```python
print(concept_id_set)
```

### knowledgeTag_skillID.txt 파일에서 id 추출
- 위에서 추출한 concept_id와 같은지 확인

```python
txt_path = './data/knowledgeTag_skillID.txt'

knowledgeTag_skillID = open(txt_path, 'r')
print(knowledgeTag_skillID)
```

```python
lines = knowledgeTag_skillID.readlines()
```

```python
lines
```

```python
id = []
new_id = []

# 각 요소에 대해
for item in lines:
    # 탭(\t)을 기준으로 문자열 분리하고
    # 첫 번째 요소를 id에, 두 번째 요소를 new_id에 추가
    parts = item.split('\t')
    id.append(parts[0])
    new_id.append(parts[1].strip())  # \n을 제거하기 위해 strip() 사용
    
print("id:", id)
print("new_id:", new_id)
```

```python
# 원소의 자료형 int로 
id = [int(i) for i in id]
```

```python
id_set = set(id)
len(id_set)
```

```python
print(concept_id_set)
```

```python
print(id_set)
```
