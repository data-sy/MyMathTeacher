```python
import json
```

```python
json_path = './data/knowledge_tags.json'
with open(json_path, 'r', encoding='utf-8') as j :
    knowledge_tags = json.load(j)
```

```python
knowledge_tags
```

```python
len(knowledge_tags)
```

```python
knowledge_tags['0']['fromConcept']['id']
```

```python
file_path = "./data/insert_knowledge_tags.sql"

with open(file_path, "w") as file:
    for knowledge_tag_id in knowledge_tags :
        from_concept_id = knowledge_tags[knowledge_tag_id]['fromConcept']['id']
        to_concept_id = knowledge_tags[knowledge_tag_id]['toConcept']['id']
        sql = f"INSERT INTO knowledge_tags values ({knowledge_tag_id}, {from_concept_id}, {to_concept_id});"
        file.write(sql + "\n")
```
