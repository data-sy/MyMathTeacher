```python
import csv
from PIL import Image, ImageDraw, ImageFont
```

```python
width, height = 400, 500
background_color = (0, 0, 0)  # 검은색

# 이미지 생성
image = Image.new("RGB", (width, height), background_color)

# 그리기 도구 생성
draw = ImageDraw.Draw(image)

# 사각형 그리기
left_top = (1, 1)
right_bottom = (width-2, height-2)
rectangle_color = (255, 255, 255)  # 흰색
draw.rectangle([left_top, right_bottom], fill=rectangle_color)

# 텍스트 그리기
text = "Hello, Python!"
font_size = 20
font_color = (0, 0, 0)  # 검은색
font = ImageFont.truetype("arial.ttf", font_size)
text_x = width//20
text_y = height//20
draw.text((text_x, text_y), text, fill=font_color, font=font)

# 이미지 저장
image.save("images/example.png")

# 이미지 표시 (선택 사항)
image.show()
```

## item별 이미지

```python
# item별 이미지

# 이미지 생성
width, height = 800, 200
background_color = (0, 0, 0)  # 검은색

# 사각형 그리기
left_top = (1, 1)
right_bottom = (width-2, height-2)
rectangle_color = (255, 255, 255)  # 흰색

# 텍스트 그리기
font_size = 10
font_color = (0, 0, 0)  # 검은색
font_path = "../GmarketSansTTF/GmarketSansTTFMedium.ttf"
font = ImageFont.truetype(font_path, font_size)
text_x = width//20
text_y = height//6

for i, diag in enumerate(split_arrs):
    for j, item in enumerate(diag):
        # 이미지 생성
        image = Image.new("RGB", (width, height), background_color)

        # 그리기 도구 생성
        draw = ImageDraw.Draw(image)

        # 사각형 그리기
        draw.rectangle([left_top, right_bottom], fill=rectangle_color)

        # 텍스트 그리기
        text = item[3]+'-'+item[4]+'-'+item[5] +'\n\n' + item[6] +'\n' + item[8]
        draw.text((text_x, text_y), text, fill=font_color, font=font)

        # 이미지 저장
        save_path = 'images/diagitems/diag'+str(i+1)+'item'+str(j+1)+'.png'
        image.save(save_path)
        
```

## 학습지 전체 만들기 최종

```python
import csv
from PIL import Image, ImageDraw, ImageFont
```

```python
csv_file_path = 'diagTest.csv'

arr = []
with open(csv_file_path, mode='r', newline='', encoding='UTF-8') as file:
    reader = csv.reader(file)
    for row in reader:
        arr.append(row)
        
arr=arr[1:]
```

```python
split_arrs=[]
for i in range(0, len(arr), 20):
    split_arrs.append(arr[i:i+20])
# split_arrs[i] : 진단평가 i+1 의 정보
```

```python
# 학습지 전체로 만들면?

# 이미지 생성
width, height = 1000, 1000
background_color = (0, 0, 0)  # 검은색

# 사각형 그리기
left_top = (1, 1)
right_bottom = (width-2, height-2)
rectangle_color = (255, 255, 255)  # 흰색

# 텍스트 그리기
font_size = 15
font_color = (0, 0, 0)  # 검은색
font_path = "../GmarketSansTTF/GmarketSansTTFMedium.ttf"
font = ImageFont.truetype(font_path, font_size)
text_x = 50

cnt=0
for i, diag in enumerate(split_arrs):
    diag01 = diag[:10]
    image = Image.new("RGB", (width, height), background_color)
    draw = ImageDraw.Draw(image)
    draw.rectangle([left_top, right_bottom], fill=rectangle_color)
    for j, item in enumerate(diag01):
        if (j<5):
            text_x = 50
            text_y = 40 + 200*j
        elif(j<10):
            text_x = 50 + 500
            text_y = 40 + 200*(j-5)
        text = item[2]+'번'+'\n\n'+item[3]+'-'+item[4]+'-'+item[5] +'\n\n' + item[6] +'\n' + item[8]
        draw.text((text_x, text_y), text, fill=font_color, font=font)
    # 이미지 저장
    save_path = 'images/diag/diag'+str(i+1)+'(1)_'+item[1]+'.png'
    image.save(save_path)
    diag02 = diag[10:]
    image = Image.new("RGB", (width, height), background_color)
    draw = ImageDraw.Draw(image)
    draw.rectangle([left_top, right_bottom], fill=rectangle_color)
    for j, item in enumerate(diag02):
        if (j<5):
            text_x = 50
            text_y = 40 + 200*j
        elif(j<10):
            text_x = 50 + 500
            text_y = 40 + 200*(j-5)
        text = item[2]+'번'+'\n\n'+item[3]+'-'+item[4]+'-'+item[5] +'\n\n' + item[6] +'\n' + item[8]
        draw.text((text_x, text_y), text, fill=font_color, font=font)
    # 이미지 저장
    save_path = 'images/diag/diag'+str(i+1)+'(2)_'+item[1]+'.png'
    image.save(save_path)
        
    # 이미지 표시
    if cnt==0:
        image.show()
        cnt = cnt+1    
```
