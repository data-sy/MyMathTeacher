```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution() #텐서플로우 v2 에서도 실행 가능하게 하기
from predict import predict
import statistics
import json
```

```python
# 데이터 증폭은 100
print("=== 증폭 100 ===")

##### 맞았을 때
# 각 인덱스에 '해당 확률'/'나머지의 평균'/그 차이/'나머지의 표준편차' 들어 있음
o_pro_list = []
o_another_avg_list = []
o_diff_list = []
o_another_std_list = []

for skill_id in range(1, 1866):
    input = [[skill_id, 1]]*100
    # ai실행
    output = predict([input])
    probability = output.pop(skill_id-1)
    # 자기 확률 챙기고
    o_pro_list.append(probability)
    # 나머지의 평균, 표준편차
    another_avg = statistics.mean(output)
    o_another_avg_list.append(another_avg)
    o_diff_list.append(probability-another_avg)
    o_another_std_list.append(statistics.stdev(output))

# 확률들 모두 모은 것의 통계치
print(f"맞혔을 때 해당 확률들 중 최댓값: {max(o_pro_list)}")
print(f"맞혔을 때 해당 확률들 중 최솟값: {min(o_pro_list)}")
print(f"맞혔을 때 해당 확률들의 평균: {statistics.mean(o_pro_list)}")
print(f"맞혔을 때 해당 확률들의 표준편차: {statistics.stdev((o_pro_list))}")

# 각 단위개념 별로 해당과 나머지의 관계 확인
# 차이가 ㅇㅇ 이상인 경우 print
list00 = []
list01 = []
list02 = []
for i, diff in enumerate(o_diff_list):
    if(diff)<0:
        list00.append([i, diff])
    if(diff)<=0.1:
        list01.append([i, diff])
    elif(diff)<=0.2:
        list02.append([i, diff])

print(f"차이 0 미만인 확률 개수 : {len(list00)}")
print(f"차이 0.1이하인 확률 개수 : {len(list01)}")
print(f"차이 0.1초과 0.2이하인 확률 개수 : {len(list02)}")
```

```python
# 확률이 0.x 미만인 리스트들
pro_lists = [[] for _ in range(11)]
for i, pro in enumerate(o_pro_list):
    index = min(int(pro * 10), 10)
    pro_lists[index].append([i, pro])
for i, pro_list in enumerate(pro_lists):
    if i < 10:
        print(f"확률이 {i / 10} 미만인 경우: {len(pro_list)}")
    else:
        print(f"확률이 0.9 이상 1 미만인 경우: {len(pro_list)}")
```

# 같은 성취수준 개념들 정오답 처리 추가해보자.
- dif_minus_id에 대해서 주변부 개념을 추가해서 dif가 준다면 의미 있어.
- 줄지 않는다면 과감히 버리자!

```python
# dif_minus_id의 주변개념들 찾기
# [스킬아이디,성취수준아이디]
d_a_list = []
for did in dif_minus_id:
    for con in concepts:
        if con['skill_id'] == did :
            achievement_id = con['achievement_id']
            d_a_list.append([did, achievement_id])
print(len(d_a_list))
for i in d_a_list:
    tmp = []
    for con in concepts:
        if con['achievement_id'] == i[1]:
            tmp.append(con['skill_id'])
    # i[2]에 같은 성취수준아이디를 갖는 스킬아이디 목록 넣기
    i.append(tmp)

# [스킬아이디, 성취수준아이디, [같은 성취수준아이디를 갖는 스킬아이디 목록]]
print(d_a_list[100])

# 딕셔너리로 바꾸기
d_a_dic = {}
for i in d_a_list:
    d_a_dic[i[0]]=i[2]
print(d_a_dic)
print(len(d_a_dic))
```

```python
# 얘는 concept 없는 애들 개수 = 즉, 버려도 돼!!
print(len(dif_minus_id)-len(d_a_dic))
```

```python
##### 맞았을 때
o_pro_list2 = []

for skill_id, ach_id_list in d_a_dic.items():
    input_data = [[skill_id, 1]]*500 + [[x, 1] for x in ach_id_list for _ in range(500)]
    # ai실행
    output = predict([input_data])
    probability = output.pop(skill_id-1)
    o_pro_list2.append([skill_id, probability])
```

```python
##### 틀렸을 때
x_pro_list2 = []

for skill_id, ach_id_list in d_a_dic.items():
    input_data = [[skill_id, 0]]*500 + [[x, 0] for x in ach_id_list for _ in range(500)]
    # ai실행
    output = predict([input_data])
    probability = output.pop(skill_id-1)
    x_pro_list2.append([skill_id, probability])
```

```python
# o_pro_list2를 딕셔너리로 수정
# skill_id가 key, proba가 value
o_pro_dic2 = {}
for i in o_pro_list2:
    o_pro_dic2[i[0]]=i[1]

# 평균 올랐는지 확인
o_pro_list2_1 = list(o_pro_dic2.values())
print(f"(맞았을 때) 평균: {statistics.mean(o_pro_list2_1)}")
print(f"(맞았을 때) 표준편차: {statistics.stdev((o_pro_list2_1))}")
```

```python
# x_pro_list2를 딕셔너리로 수정
# skill_id가 key, proba가 value
x_pro_dic2 = {}
for i in x_pro_list2:
    x_pro_dic2[i[0]]=i[1]

# 평균 내렸는지 확인
x_pro_list2_1 = list(x_pro_dic2.values())
print(f"(틀렸을 때) 평균: {statistics.mean(x_pro_list2_1)}")
print(f"(틀렸을 때) 표준편차: {statistics.stdev((x_pro_list2_1))}")
```

```python
arr = list(o_pro_dic2.keys())
result_fin = []
# 맞았을 때의 새확률-기존확률<0 인 경우 : 적으면 관계성이 유의미하다? 얘는 그냥 해본 거. 중요한 건 아냐
cnt1=0
# 새확률의 맞았을때-틀렸을때<0 인 경우 : 얘는 적어야 해. 많으면 이 ai가 해당 개념에 대해서는 써먹을 방법이 없다는 결론에 도달 ㅜㅠ
cnt2=0
for i in arr:
    # 기존 확률, 맞았을 때
    a = o_pro_list[i-1]
    # 새 확률, 맞았을 때
    b = o_pro_dic2[i]
    # 기존 확률, 틀렸을 때
    c = x_pro_list[i-1]
    # 새 확률, 틀렸을 때
    d = x_pro_dic2[i]
    if b-a<0:
        cnt1+=1
    if b-d<0:
        cnt2+=1
print(cnt1)
print(cnt2)
# 아까 390개에서 242개로 줄긴 줄었음...
```
