```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution() #텐서플로우 v2 에서도 실행 가능하게 하기
from predict import predict
import statistics
import json
```

```python
# concepts_all 파일
json_path = '../Analysis, Design/analysis_data/data/concepts_all.json'
with open(json_path, 'r', encoding='utf-8') as j :
    concepts = json.load(j)
```

```python
concepts[0]
```

```python
# 초등 개념만 추출
ele_skill_id_list = []
for con in concepts:
    if con['school_level']=='초등':
        ele_skill_id_list.append(con['skill_id'])
print(len(ele_skill_id_list))
```

```python
# 중등 개념만 추출
mid_skill_id_list = []
for con in concepts:
    if con['school_level']=='중등':
        mid_skill_id_list.append(con['skill_id'])
print(len(mid_skill_id_list))
```

```python
# 중등 문제만 가지고 테스트
o_pro_list = []
o_id_pro_dic = {}
for skill_id in mid_skill_id_list:
    input_data = [[skill_id, 1]]*500 # 데이터 증폭 500
    # ai실행
    output = predict([input_data])
    probability = output.pop(skill_id-1)
    # 확률 목록
    o_pro_list.append(probability)
    o_id_pro_dic[skill_id]=probability
```

```python
# 확률들 모두 모은 것의 통계치
print(f"맞혔을 때 해당 확률들 중 최댓값: {max(o_pro_list)}")
print(f"맞혔을 때 해당 확률들 중 최솟값: {min(o_pro_list)}")
print(f"맞혔을 때 해당 확률들의 평균: {statistics.mean(o_pro_list)}")
print(f"맞혔을 때 해당 확률들의 표준편차: {statistics.stdev((o_pro_list))}")

# 확률이 0.x 미만인 리스트들
pro_lists = [[] for _ in range(11)]
for i, pro in o_id_pro_dic.items():
    index = min(int(pro * 10), 10)
    pro_lists[index].append([i+1, pro])
for i, pro_list in enumerate(pro_lists):
    if i < 10:
        print(f"확률이 {(i+1) / 10} 미만인 경우: {len(pro_list)}")
```
